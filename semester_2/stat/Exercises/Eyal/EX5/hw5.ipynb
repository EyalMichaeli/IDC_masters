```python
winery_counts = raw_df.winery.value_counts()

wineries_more_than_1 = winery_counts[winery_counts > 1].index

df = raw_df[raw_df.winery.isin(wineries_more_than_1)]
print(f"We removed {len(raw_df) - len(df)} rows from the data\n")

df.sample(5)
```

Re-run peoblem 4 (normal p-values)

```python
# for later use

variable = 'points'

def ssquares(x):
    return np.sum((x - np.mean(x)) ** 2)

ss_wit = df.groupby('winery')[variable].agg(ssquares).sum()
ss_tot = np.sum((df[variable] -  df[variable].mean()) ** 2)

n = len(df)
k = len(wineries_more_than_1)

ss_bet = ss_tot - ss_wit

MSE_bet = ss_bet / (k-1)
MSE_wit = ss_wit / (n-k)
```

```python
EPSILON = 10 ** -6

def two_sample_t(data1, data2):
    data1_mean = np.mean(data1)
    data2_mean = np.mean(data2)
    
    n0 = len(data1)
    n1 = len(data2)
    dof = n0 + n1 - 2
    
    s = np.sqrt((np.sum((data1-data1_mean)**2) + np.sum((data2-data2_mean)**2)) / dof)
    
    s = max(EPSILON, s)
    
    t_statistic = (data2_mean - data1_mean) / (s * np.sqrt(1 / n0 + 1 / n1))
    p_val = stats.t.sf(abs(t_statistic), dof)

    return t_statistic, p_val, dof


wineries_pairs = [(winery_1, winery_2) for i, winery_1 in enumerate(wineries_more_than_1) for winery_2 in wineries_more_than_1[i + 1: ]]
print(f"There are {len(wineries_pairs)} t-tests to be done")

tests_pvals = list()
for winery_1, winery_2 in wineries_pairs:
    data1 = df[df.winery == winery_1]["points"]
    data2 = df[df.winery == winery_2]["points"]
    
    t_stat, p_val, dof = two_sample_t(data1, data2)

    tests_pvals.append(p_val)


ALPHA = 0.05
p_vals_significant = [p_val for p_val in tests_pvals if p_val < ALPHA / 2] 


print(f"\nThere are {len(p_vals_significant)} out of {len(wineries_pairs)} total p-values that with a p-value lower than {ALPHA}")

print(f"\nUnder null hypothisis we expect to reject {len(wineries_pairs) * ALPHA // 1}, and we rejected {len(p_vals_significant)}.")
```

## Problem 3 (Exact size of Bonferroni's test)
Suppose that we run multiple tests with independent data and obtain P-values $p_1,\ldots,p_n$. We wish to test the null hypothesis:
$$
H_0\,:\,\text{All tests are null}
$$
at the level $\alpha$ (e.g., $\alpha=0.05$). In class, we introduced Bonferroni's procedure which is equivalent to: Reject $H_0$ if $\min p_i \leq \alpha/n$. 
1. Show that the  size of the test in Bonferroni's procedure is at most $\alpha$, regardless if the hypotheses are independent or not. 
2. Assuming that the hypotheses are independent, find the exact size of the test in Bonferroni's procedure. 
3. For $\alpha=0.05$, evalaute the difference between $\alpha$ and the exact test's size for $n=2,...,50$. Discuss what you see. 



1. Show that the  size of the test in Bonferroni's procedure is at most $\alpha$, regardless if the hypotheses are independent or not. 

taken from lecture 7:
<br>
We know that the boferroni correction is simply replacing $\alpha$ with $\frac{\alpha}{m}$ where m is the number of tests

$$
P(H_0 rejection | H_0)  = P(MIN(p_i) \leq \frac{\alpha}{n}) 
$$

$$
= P( p_1 \leq \frac{\alpha}{n} \cup ... \cup p_n \leq \frac{\alpha}{n})
$$

Because this is a union, it's less than or equal to:

$$
\leq \sum_{i=1}^{n} P( p_i \leq \frac{\alpha}{n}) 
$$

$$
= \sum_{i=1}^{n} \frac{\alpha}{n} = \frac{\alpha}{n} \cdot \sum_{i=1}^{n} 1 = \frac{\alpha}{n} \cdot n  = \alpha
$$

2. Assuming that the hypotheses are independent, find the exact size of the test in Bonferroni's procedure. 

$$
P(H_0 rejection | H_0)  = P(MIN(p_i) \leq \frac{\alpha}{n}) 
$$

$$
= P( p_1 \leq \frac{\alpha}{n} \cup ... \cup p_n \leq \frac{\alpha}{n})
$$

$$
1 - P( p_1 > \frac{\alpha}{n} \cap ... \cap p_n > \frac{\alpha}{n})
$$

and because of the independence of the t tests (given):

$$
= 1 - \prod_{i=1}^{n} P( p_i > \frac{\alpha}{n}) = 1 - \prod_{i=1}^{n} ( 1 - P( p_i \leq \frac{\alpha}{n} )) = 1 - \prod_{i=1}^{n} ( 1 - \frac{\alpha}{n}) 
= 1 - ( 1 - \frac{\alpha}{n} )^n
$$

If we remember Euler from calculus class, then we should remember that when n is very large (goes to infinity), the above expression will go to $1 - e^{-\alpha}$

3. For $\alpha=0.05$, evalaute the difference between $\alpha$ and the exact test's size for $n=2,...,50$. Discuss what you see. 

```python
ALPHA = 0.05
n_array = np.arange(2, 50+1)

exact_size_per_n = np.array([1 - (1 - ALPHA / n) ** n for n in n_array])
diffs = np.ones_like(exact_size_per_n) * ALPHA - exact_size_per_n
diffs.shape
```

```python
# plot

plt.plot(n_array, np.ones_like(n_array) * ALPHA, label="ALPHA")
plt.plot(n_array, exact_size_per_n, label="Exact Size")
plt.title("ALPHA & Exact Size by n (number of t-tests)")
plt.legend()

plt.show()
```

```python
plt.plot(n_array, diffs, label="alpha - exact_size(diff)")

plt.title("ALPHA - Exact Size (diff) by n (number of t-tests)")
plt.show()
```

Firstly, we see that the exact size decreases with n, the more t tests, the lower the exact size. We also see that the exact size goes to $1 - e^{-\alpha}$ as suspected, when n gets larger.

As a result by the fact alpha is a constant, we see that the difference gets bigger and bigger, the same way that the exact size get smaller and smaller.
